```python
print("\nAnalizamos la base de datos:")

print("\nDescripción general del dataset:")
print(data.info())
```

```python
print("\nEstadísticas descriptivas:")
print(data.describe())
```

```python
print("\nContar valores nulos por columna:")
print(data.isnull().sum())
```

```python
print("\nEliminar las filas donde el valor de la educacion de madre y padre sea 99:")
data = data[data['educacion_madre'] != 99]
data = data[data['educacion_padre'] != 99]
print(data.describe())
```

```python
!git checkout Análisis_descriptivo
!git checkout main
!git merge Análisis_descriptivo
```

```python
!git branch Limpieza_de_datos
```

```python
#Cambiar nulos por la media de su respectiva columna
datos_numericos = data.select_dtypes(include=["number"])
```

```python
print("\nRellenar valores nulos en columnas numéricas con el promedio")
datos_numericos.fillna(datos_numericos.mean(), inplace=True)
data.update(datos_numericos)
print(data.isnull().sum())
```

```python
print("\nTranformar genero y etnia a dummy")
data = pd.get_dummies(data, columns=['genero'], drop_first=True)
data= pd.get_dummies(data, columns=['etnia'], drop_first=True)
```

```python
print("\nHistograma")
data.hist(figsize=(12, 8))
plt.suptitle("Distribuciones de las variables")
plt.show()
```

```python
print("\nMatriz de correlación")
plt.figure(figsize=(10, 6))
sns.heatmap(data.corr(), annot=True, cmap="coolwarm")
plt.title("Matriz de Correlación")
plt.show()
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
#Escalar datos
print("\nEscalar los datos numéricos")
escalador = StandardScaler()
data_escalados = escalador.fit_transform(data.select_dtypes(include=["number"]))

#Aplicar PCA con 2 componentes principales
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_escalados)

# Mostrar la varianza explicada
print("\nVarianza explicada por cada componente principal:")
print(pca.explained_variance_ratio_)

print("\nVisualizar los datos proyectados")
plt.figure(figsize=(8, 6))
plt.scatter(data_pca[:, 0], data_pca[:, 1], alpha=0.7)
plt.title("Proyección PCA")
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.show()
```
