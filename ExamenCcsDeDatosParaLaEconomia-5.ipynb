```python
import pandas as pd

# Cargar el archivo de texto
archivo = "ingresos-1.txt"
df = pd.read_csv(archivo, sep="\t", encoding="utf-8")  # Ajusta el separador si es necesario

# Mostrar las primeras filas
print("Primeras filas del archivo:")
print(df.head())
```

```python
# Información general del dataset
print("\nInformación del dataset:")
df.info()

# Resumen estadístico de las columnas numéricas
print("\nResumen estadístico:")
print(df.describe())

# Verificar valores nulos
print("\nValores nulos por columna:")
print(df.isnull().sum())
```

```python
print("\nEliminar las filas donde el valor de la educacion de madre y padre sea 99:")
df = df[df['educacion_madre'] != 99]
df = df[df['educacion_padre'] != 99]
print(df.describe())
```

```python
#Cambiar nulos por la media de su respectiva columna
df_numericos = df.select_dtypes(include=["number"])
print("\nRellenar valores nulos en columnas numéricas con el promedio")
df_numericos.fillna(df_numericos.mean(), inplace=True)
df.update(df_numericos)
print(df.isnull().sum())
print("\nTranformar genero y etnia a dummy")
df = pd.get_dummies(df, columns=['genero'], drop_first=True)
df= pd.get_dummies(df, columns=['etnia'], drop_first=True)
```

```python
# Resumen de ingresos
print("Resumen de ingresos:")
print(df['ingresos'].describe())

# Media, mediana y desviación estándar
media_ingresos = df['ingresos'].mean()
mediana_ingresos = df['ingresos'].median()
std_ingresos = df['ingresos'].std()

print(f"Media de ingresos: {media_ingresos}")
print(f"Mediana de ingresos: {mediana_ingresos}")
print(f"Desviación estándar de ingresos: {std_ingresos}")
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
#Escalar datos
print("\nEscalar los datos numéricos")
escalador = StandardScaler()
data_escalados = escalador.fit_transform(df.select_dtypes(include=["number"]))

#Aplicar PCA con 2 componentes principales
pca = PCA(n_components=2)
data_pca = pca.fit_transform(data_escalados)

# Mostrar la varianza explicada
print("\nVarianza explicada por cada componente principal:")
print(pca.explained_variance_ratio_)

print("\nVisualizar los datos proyectados")
plt.figure(figsize=(8, 6))
plt.scatter(data_pca[:, 0], data_pca[:, 1], alpha=0.7)
plt.title("Proyección PCA")
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.show()
```

```python
#Supuestos del PCA y su cumplimiento
#1.⁠ ⁠Linealidad
#El PCA asume que la relación entre las variables es lineal.
#Por qué se cumple: La dispersión de los datos en los componentes principales sugiere que la relación entre las variables 
#originales es principalmente lineal.

#2.⁠ ⁠Escalamiento (Normalización de los datos)
#Las variables deben estar en la misma escala, ya que, el PCA es sensible a la magnitud y unidades de las variables.
#Por qué se cumple: Se aplicó StandardScaler para escalar los datos, asegurando que cada variable tenga una media de 0 y 
#desviación estándar de 1. Garantizando que se cumpla este supuesto.

#3.⁠ ⁠No colinealidad perfecta
#El PCA requiere que las variables no estén perfectamente correlacionadas para evitar redundancia. 
#Por qué se cumple: Al revisar la matriz de correlación, no se detectaron correlaciones perfectas entre las variables originales, 
#lo cual garantiza que el PCA puede encontrar componentes principales únicos.

#4.⁠ ⁠Varianza máxima
#El PCA busca capturar la mayor cantidad de varianza en los primeros componentes principales.
#Por qué se cumple: Los valores de explained_variance_ratio_ indican que los dos primeros componentes capturan una 
#gran proporción de la varianza total de los datos.
```
